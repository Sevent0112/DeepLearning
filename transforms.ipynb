```python
from torch.utils.data import Dataset
from torchvision import transforms
from PIL import Image
```

# 常见transforms用法

```python
img = Image.open("data/16838648_415acd9e3f.jpg")
```

```python
print(img)
```

ToTensor的使用

```python
trans_totensor = transforms.ToTensor()
img_tensor = trans_totensor(img)
img_tensor
```

# Normalize

```python
trans_norm = transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5]) #图片三通道
img_norm = trans_norm(img_tensor)
```

```python
img_norm
```

# resize

```python
print(img.size)
```

```python
trans_resize = transforms.Resize((512, 512))
print(trans_resize(img_tensor))
```

# compose

```python
trans_resize_2 = transforms.Resize(512)
```

```python
trans_compose = transforms.Compose([trans_resize_2, trans_totensor])
```

```python
trans_compose(img)
```

```python
# randomcrop
```

```python
trans_crop = transforms.RandomCrop(320)
trans_crop(img)
```
